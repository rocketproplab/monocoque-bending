```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numSteps = 4

SFy = 1.25
SFu = 1.5 # factor of safety, ultimate

# material properties
rho = 0.0975 # lb/in^3
sigmay = 19000 # psi
sigmau = 33000 # psi

sigmastar = min(sigmay/SFy, sigmau/SFu)

# airframe calculations
length = 36 # length of monocoque section
radius_mean = np.linspace(3,6) # in
thickness = np.linspace(0.125, .5, numSteps)

plt.figure(1)
plt.grid(True)
plt.title('Moment vs. Radius Mean')
plt.xlabel('Mean Radius (in)')
plt.ylabel('Moment Allowable (lb-in)')
plt.legend(['0.125', '0.25', '0.375', '0.5'])

plt.figure(2)
plt.grid(True)
plt.title('Moment vs. Mass')
plt.xlabel('Mass (lbm)')
plt.ylabel('Moment Allowable (lb-in)')


for thick in thickness:

    circumference = 2*np.pi*radius_mean
    mass = circumference * thick * length * rho # mass of cylinder
    # moment of inertia of the cylinder
    inertia = mass * radius_mean**2

    # calculating stresses
    #sigma = moment * y / I
    moment = sigmastar * inertia / length
    
    plt.figure(1)
    plt.plot(radius_mean, moment)
    
    plt.figure(2)
    plt.plot(mass, moment)

plt.figure(1)
plt.legend(['0.125', '0.25', '0.375', '0.5'])
plt.figure(2)
plt.legend(['0.125', '0.25', '0.375', '0.5'])
plt.show()
```
